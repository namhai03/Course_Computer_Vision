```python
import cv2
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Image
```

## I. Display Image

```python
url = "/content/images/girl-gray.png"
```

```python
Image(filename = url)
```

## II. Reading Images using OpenCV

### 1. Read Image and Display level Image data

```python
bw_img = cv2.imread(url, cv2.IMREAD_GRAYSCALE)
print(bw_img)
```

```python
print("Shape:", bw_img.shape)
print("Data type:", bw_img.dtype)
```

### 2. Image Coordiantes

```python
print(bw_img[100,200])
print(bw_img[100][200])
```

## III. Display using Matplotlib

```python
cat = cv2.imread("/content/images/cat.jpg", cv2.IMREAD_GRAYSCALE)
plt.imshow(cat, cmap="gray")
```

```python
print(cat)
```

## IV. Save image

```python
image = cv2.imread("/content/images/cat.jpg", cv2.IMREAD_GRAYSCALE)
plt.figure(figsize=[7,7])
plt.imshow(image, cmap="gray")
plt.title("Cat")
```

```python
cv2.imwrite("/content/images/Cat-gray.png", image)
```

```python
image = cv2.imread("/content/images/Cat-gray.png")
plt.figure(figsize=[7,7])
plt.imshow(image)
plt.title("Cat saved")
```
