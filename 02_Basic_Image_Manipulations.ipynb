```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
from IPython.display import Image
plt.rcParams["image.cmap"] = "gray"
```

# 1. Pixel Operations

```python
number = cv2.imread("18.png", cv2.IMREAD_GRAYSCALE)
plt.figure(figsize=[5,5])
plt.imshow(number)
```

```python
print(number[10,10])
print(number[0,0])
print(number[9,16])
```

```python
number_copy = number.copy()
number_copy = number_copy[9:22,4:22]
```

```python
number_copy[0,0] = 0
number_copy[5,14:16] = 255
plt.figure(figsize=[10,10])
plt.xticks(range(0, number_copy.shape[1], 1))
plt.yticks(range(0, number_copy.shape[0], 1))
plt.imshow(number_copy)
```

```python
print(number_copy)
```

```python
# flip 1, 0, -1
plt.subplot(141); plt.imshow(cv2.flip(number, 1))
plt.subplot(142); plt.imshow(cv2.flip(number, 0))
plt.subplot(143); plt.imshow(cv2.flip(number, -1))
plt.subplot(144); plt.imshow(number)
```
