# Split and Merge Color

```python
cat_bgr = cv2.imread("cat.jpg", cv2.IMREAD_COLOR)
b, g, r = cv2.split(cat_bgr)
plt.figure(figsize = [20,20])
plt.subplot(141); plt.imshow(r); plt.title("Red channel")
plt.subplot(142); plt.imshow(g); plt.title("Green channel")
plt.subplot(143); plt.imshow(b); plt.title("Blue channel")

cat_merge = cv2.merge((r,g,b))
plt.subplot(144); plt.imshow(cat_merge); plt.title("Merge Output")
```

```python
cat_hsv = cv2.cvtColor(cat_bgr, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(cat_hsv)
plt.figure(figsize = [20,20])
plt.subplot(141); plt.imshow(h); plt.title("Hue channel")
plt.subplot(142); plt.imshow(s); plt.title("Saturation channel")
plt.subplot(143); plt.imshow(v); plt.title("Value channel")
plt.subplot(144); plt.imshow(cat_bgr[:,:,::-1]); plt.title("Original Image")
```

```python
h_new = h - 10
cat_hsv_merge = cv2.merge((h_new,s,v))
cat_rgb_merge = cv2.cvtColor(cat_hsv_merge, cv2.COLOR_HSV2RGB)

plt.figure(figsize = [20,20])
plt.subplot(141); plt.imshow(h_new); plt.title("Hue channel")
plt.subplot(142); plt.imshow(s); plt.title("Saturation channel")
plt.subplot(143); plt.imshow(v); plt.title("Value channel")
plt.subplot(144); plt.imshow(cat_rgb_merge); plt.title("Original Image")
```
