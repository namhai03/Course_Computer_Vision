```python
import numpy as np
import cv2
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
```

# 1. Read gray scale imaage as "color"

```python
girl = cv2.imread("girl-gray.png", cv2.IMREAD_COLOR)

plt.figure(figsize=[5,5])
plt.imshow(girl)
print(girl.shape)
```

```python
yellow = (0, 255, 255)
red = (0, 0, 255)
maagenta = (255, 0, 255)
green = (0, 255, 0)
```

# 2. Drawing a line

```python
girl_line = girl.copy()
girl_line = cv2.line(girl_line, (100,100), (137,105), red[::-1], thickness=1, lineType=cv2.LINE_AA)
girl_line = cv2.line(girl_line, (150,105), (185,100), red[::-1], thickness=1, lineType=cv2.LINE_AA)
plt.figure(figsize=[7,7])
plt.imshow(girl_line)
```

```python
girl_line = girl.copy()
girl_line = cv2.line(girl_line, (100,100), (137,105), red[::-1], thickness=1, lineType=cv2.LINE_8)
girl_line = cv2.line(girl_line, (150,105), (185,100), red[::-1], thickness=1, lineType=cv2.LINE_8)
plt.figure(figsize=[7,7])
plt.imshow(girl_line)
```

# 3. Drawing a circle

```python
girl_circle = girl.copy()
girl_circle = cv2.circle(girl_circle, (140,120), 100, red[::-1], thickness=2, lineType=cv2.LINE_AA)
plt.figure(figsize=[7,7])
plt.imshow(girl_circle)
```

# 4. Drawing rectangle image

```python
girl_rectangle = girl.copy()
girl_rectangle = cv2.rectangle(girl_rectangle, (60,40), (210,230), red[::-1], thickness=3, lineType=cv2.LINE_AA)
plt.figure(figsize=[7,7])
plt.imshow(girl_rectangle)
```

# 5. Adding Text

```python
girl_text = girl.copy()
text = "Laddy twenty-fine"
font_face = cv2.FONT_HERSHEY_SIMPLEX
font_scale = 0.5
font_color = red[::-1]
font_thickness = 1
girt_text = cv2.putText(girl_text, text, (80,350), font_face, font_scale, font_color, font_thickness, lineType=cv2.LINE_AA)
plt.imshow(girl_text)
```
