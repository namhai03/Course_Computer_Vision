```python
import os

folder_path = '/content/data_train/train/labels'  # Thay th·∫ø ƒë∆∞·ªùng d·∫´n t·ªõi th∆∞ m·ª•c c·ªßa b·∫°n
search_str = '15'
replace_str = '0'

def replace_first_15(file_path):
    with open(file_path, 'r') as file:
        content = file.read()

    content = content.replace(search_str, replace_str, 1)

    with open(file_path, 'w') as file:
        file.write(content)

def process_files_in_folder(folder_path):
    for filename in os.listdir(folder_path):
        if filename.endswith('.txt'):
            file_path = os.path.join(folder_path, filename)
            replace_first_15(file_path)

if os.path.exists(folder_path):
    process_files_in_folder(folder_path)
    print(f'ƒê√£ th·ª±c hi·ªán thay th·∫ø trong t·∫•t c·∫£ c√°c file .txt trong th∆∞ m·ª•c: {folder_path}')
else:
    print(f'Th∆∞ m·ª•c kh√¥ng t·ªìn t·∫°i: {folder_path}')
```

```python
from ultralytics import YOLO
model = YOLO('yolov8n.pt')
```

```python
import yaml
file_path = '/content/data_train/train/images'
# T·∫°o d·ªØ li·ªáu YAML
data = {
    'name': ["0"],
    'nc': 1,
    'train': file_path,
    'val': file_path,
}

# Ghi d·ªØ li·ªáu v√†o file YAML
with open('/content/data.yaml', 'w') as yaml_file:
    yaml.dump(data, yaml_file, default_flow_style=False)

print("ƒê√£ t·∫°o file data.yaml th√†nh c√¥ng.")
```

```python
results = model.train(data='/content/data.yaml', epochs=50, imgsz=800)
```
